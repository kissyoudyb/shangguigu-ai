## 第一节 分词模式

```python
import jieba
```

```python
text = "小明毕业于北京大学计算机系"
```

```python
# 1.精确模式
jieba.lcut(text)
```

```python
# 2.全模式
jieba.lcut(text,cut_all=True)
```

```python
# 3.搜索引擎模式
jieba.lcut_for_search(text)
```
